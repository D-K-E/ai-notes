```python
# packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image
img_n = cv2.imread("images/linesNoter.jpg")
img_h = cv2.imread("images/linel.png")

image1 = np.copy(img_n)
image2 = np.copy(img_h)

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,20))

ax1.imshow(image1)
ax2.imshow(image2)
```

```python
# perform edge detection
image1_g = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)
image2_g = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)

# define trashold for the images

# image1
low1 = 250
high1 = 254

# image2
low2 = 50
high2 = 254

# Apply canny to both images

edges1 = cv2.Canny(image1_g, low1, high1)
edges2 = cv2.Canny(image2_g, low2, high2)


fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,20))

ax1.imshow(edges1, cmap='gray')
ax2.imshow(edges2, cmap='gray')
```

```python
# Define the Hough transform parameters
rho = 20
theta = np.pi/180
threshold = 80
min_line_length = 200
max_line_gap = 20

# Find lines with hough transform
lines1 = cv2.HoughLinesP(edges1, rho, theta, 
                         threshold, np.array([]),
                         min_line_length, max_line_gap
                        )
#
lines2 = cv2.HoughLinesP(edges2, rho, theta, 
                         threshold, np.array([]),
                         min_line_length, max_line_gap
                        )
#
line_image1 = np.copy(image1)
line_image2 = np.copy(image2)

for line in lines1:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image1, 
                 (x1,y1), 
                 (x2,y2), 
                 (255,0,0),5)
        
for line in lines2:
    for x1,y1,x2,y2 in line:
        cv2.line(line_image2, 
                 (x1,y1), 
                 (x2,y2), 
                 (255,0,0),5)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,20))

ax1.imshow(line_image1, cmap='gray')
ax2.imshow(line_image2, cmap='gray')
```
