```python
# packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image
img = cv2.imread('images/imageContours.jpg')

image = np.copy(img)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

```python
# convert image to grayscale
image_gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Binarize the image

ret, image_bin = cv2.threshold(image_gray, # original image
                          250, # below which everything is black
                          255, # value for those that are above threshold
                          cv2.THRESH_BINARY_INV)

plt.imshow(image_bin, cmap='gray')
```

```python
# find contours from binary image
retval, contours, hierarchy = cv2.findContours(image_bin,
                                              cv2.RETR_TREE, #contour retrieval mode
                                              cv2.CHAIN_APPROX_SIMPLE 
                                               # contour approximation method
                                              )
contour_cp = np.copy(img)

# Lets draw the contours around the image

contour_img = cv2.drawContours(contour_cp, # src image to draw on the contours
                              contours, # the contours to be drawn
                              -1, # this means all the contours from the contours 
                               # above
                              (0,255,0), # the color of the contours, here green.
                              2 # the size of the line, here 2 pixels
                              )
plt.imshow(contour_img)
```

```python
# contours on text line
img2 = cv2.imread('images/letter.tif')

image2 = np.copy(img2)

image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)

plt.imshow(image2)
```

```python
# convert image to grayscale
image_gray2 = cv2.cvtColor(image2, cv2.COLOR_RGB2GRAY)

# Binarize the image

image_gray2 = cv2.bitwise_not(image_gray2)

ret, image_bin2 = cv2.threshold(image_gray2, # original image
                          127, # below which everything is black
                          255, # value for those that are above threshold
                          cv2.THRESH_BINARY)

plt.imshow(image_bin2, cmap='gray')
```

```python
# find contours from binary image
retval, contours2, hierarchy = cv2.findContours(image_bin2,
                                              cv2.RETR_TREE, #contour retrieval mode
                                              cv2.CHAIN_APPROX_SIMPLE 
                                               # contour approximation method
                                              )
contour_cp2 = np.copy(img2)

# Lets draw the contours around the image

contour_img2 = cv2.drawContours(contour_cp2, # src image to draw on the contours
                              contours2, # the contours to be drawn
                              -1, # this means all the contours from the contours 
                               # above
                              (0,255,0), # the color of the contours, here green.
                              1 # the size of the line, here 2 pixels
                              )
plt.imshow(contour_img2)
```

```python
# get bounding boxes
%matplotlib qt

def getBboxes(contours):
    "Gets the bounding box for contour"
    b_boxes = []
    for con in contours:
        # compute b box
        x,y, width, height = cv2.boundingRect(con)
        b_box = (x,y,width,height)
        b_boxes.append(b_box)
    #
    return b_boxes

b_boxes_latin = getBboxes(contours3)

chars = []

for b_box in b_boxes_latin:
    x,y, width, height = b_box
    box_img = image3[y:y+height+4, x:x+width+4]
    chars.append(box_img)

#fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(10,20))

print(len(chars))

#ax3
plt.imshow(chars[128])
```
