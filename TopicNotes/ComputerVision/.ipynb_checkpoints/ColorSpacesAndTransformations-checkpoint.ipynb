```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
import cv2
%matplotlib inline
```

```python
# read the image

image1 = cv2.imread('images/multicolor.jpg')

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

image = np.copy(image1)

plt.imshow(image)
```

```python
# isolate channels
red = image[:,:,0]
green = image[:,:,1]
blue = image[:,:,2]

blue[:10]
```

```python
# plot the channels in grayscale

figure, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Red', color="white")
ax1.imshow(red, cmap='gray')

ax2.set_title('Green', color="white")
ax2.imshow(green, cmap='gray')

ax3.set_title('Blue', color="white")
ax3.imshow(blue, cmap='gray')
```

```python
# convert the color space from RGB to HSV
hsv_image = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)

plt.imshow(hsv_image)
```

```python
# Isolating hsv channels
hue = hsv_image[:,:,0]
saturation = hsv_image[:,:,1]
value = hsv_image[:,:,2]

# plot the channels in grayscale

figure, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20,10))

ax1.set_title('Hue', color="white")
ax1.imshow(hue, cmap='gray')

ax2.set_title('Saturation', color="white")
ax2.imshow(saturation, cmap='gray')

ax3.set_title('Value', color="white")
ax3.imshow(value, cmap='gray')
```

```python
# Define rgb and hue masks

# RGB range
lower_green = np.array([0,110,0])
upper_green = np.array([110,255,100])
```

```python
# Creating masks

# RGB mask
mask_rgb = cv2.inRange(image, lower_green, upper_green)

masked_image = np.copy(image)

masked_image[mask_rgb==0] = [0,0,0]

plt.imshow(masked_image)
```

```python
# Hue range
lower_hue = np.array([0,120,0])
upper_hue = np.array([130,180,110])
```

```python
# Creating masks

# Hue mask
mask_hue = cv2.inRange(image, lower_hue, upper_hue)

masked_image_hue = np.copy(image)

masked_image_hue[mask_hue==0] = [0,0,0]

plt.imshow(masked_image_hue)
```
