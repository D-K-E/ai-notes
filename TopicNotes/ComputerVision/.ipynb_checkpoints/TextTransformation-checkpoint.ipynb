```python
# packages
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
# was inline before see section coming up with coordinates
```

```python
# read image
image1 = cv2.imread("images/perspective.jpg")

image = np.copy(image1)

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image)
```

#### Coming Up with coordinates

Now for this example we are going to manually define the original
coordinates, and the coordinates in which we would like to see the image

For this manual definition we need to redefine our `%matplotlib inline` to `%matplotlib qt`

```python
# Marking coordinates in the image
%matplotlib inline
plt.imshow(image)
plt.plot(355,539, '.') #top left corner
plt.plot(604,644, '.') #bottom left corner
plt.plot(894,401, '.') #bottom right corner
plt.plot(657,314, '.') #top right corner
```

```python
# Define a function to do all these steps

def warpImage(img: np.ndarray([], dtype="uint8")):
    "Warps an image to a perspective"
    # Source image points
    source_pts = np.float32([
        [355, 539],
        [604,644],
        [894,401],
        [657,314]
    ])
    # target points
    target_pts = np.float32([
        [100, 200], # top left corner
        [100, 550], # bottom left corner
        [800, 550], # bottom right corner
        [800, 200] # top right corner
        
    ])
    # get perspective matrix
    pers_m = cv2.getPerspectiveTransform(source_pts, target_pts)
    # get the inverse matrix
    pers_m_inv = cv2.getPerspectiveTransform(target_pts, source_pts)
    # get image size
    im_size = (img.shape[1], img.shape[0]) # column, row
    # Compute and return warped image
    warped = cv2.warpPerspective(img, pers_m, im_size, 
                                 flags=cv2.INTER_LINEAR)
    #
    return warped
```

```python
# let's compare both the original and warped image side by side
warped_image = warpImage(img=image)

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.set_title("Source Image", color="white")
ax1.imshow(image)
ax2.set_title("Warpped Image", color="white")
ax2.imshow(warped_image)
```
