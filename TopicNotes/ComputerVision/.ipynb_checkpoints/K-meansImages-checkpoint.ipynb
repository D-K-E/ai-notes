```python
# packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image
img_h = cv2.imread("images/linel.png")

image1 = np.copy(img_h)

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

plt.imshow(image1)
```

```python
# prepare data for k-means clustering
# reshaping the data to obtain a 2d array
# pixels and 3 color values
# rows are now number of pixels,
# columns are 3 color values
pixel_vals = image1.reshape((-1,3))
#
print("new shape", pixel_vals.shape)
pixel_vals[:5]
# change the data type to float
pixel_vals =  np.float32(pixel_vals)
```

```python
# perform k-means clustering

# define stop criteria
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER,
           10, # number of max iterations allowed
           1 # minimum measure of movement signaling the convergence
            # if the movement is below this level we consider the algorithm
            # as converged
           )

k=5
ret_value, labels, centers = cv2.kmeans(pixel_vals, # data in float form 
                                        k, # number of centers 
                                        None, # any labels we want
                                        criteria, # our stop criteria that should 
                                        # be defined
                                        10, # number of attempts,
                                        # we can think of it as number of updates
                                        # on center points
                                        cv2.KMEANS_RANDOM_CENTERS
                                        # how to initialize the center points
                                       )
# displaying clusters

# Convert data into images

centers = np.uint8(centers)

segmented_data = centers[labels.flatten()]

# reshape data into original dimension

segmented_image = segmented_data.reshape(image1.shape)
labels_reshape = labels.reshape(image1.shape[0], image1.shape[1])

# plot the segmented/clustered image

plt.imshow(segmented_image)
```

```python
# Visualize a single segment

plt.imshow(labels_reshape==0, cmap='gray')
```
