```python
# packages
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# read the image
img = cv2.imread('images/edgefilter.jpg')

image1 = np.copy(img)

image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)

plt.imshow(image1)
```

### High Pass Filters

```python
# convert the image to grayscale
image = cv2.cvtColor(image1, cv2.COLOR_RGB2GRAY)

plt.imshow(image, cmap='gray')
```

```python
# Create custom kernel for detecting
# vertical edges
# c_kernel = np.array([ 
# we are using a 3x3 matrix 
# first we define an empty kernel
#     [,,]
#     [,,]
#     [,,]
# ])

# then I put negative values on the left side
# c_kernel = np.array([ 
#     [-1,,]
#     [-2,,]
#     [-1,,]
# ])
# As you can see the corners are less important


# then I put 0 on the center since I don't want the
# center to be a determining factor
# c_kernel = np.array([ 
#     [-1, 0, ]
#     [-2, 0, ]
#     [-1, 0, ]
# ])


# then I put positive values on the right side
# c_kernel = np.array([ 
#     [-1,0,1]
#     [-2,0,2]
#     [-1,0,1]
# ])
# This is actually a pretty popular filter called
# sobel filter

conv_kernel_x = np.array([ # we are using a 3x3 matrix
    [-1,0,1],
    [-2,0,2],
    [-1,0,1],
])

filtered = cv2.filter2D(image, # input image
                        -1, # type of output image -1 means the same type as 
                        # the input
                        conv_kernel_x # the kernel we created
                       )

plt.imshow(filtered, cmap='gray')
```

```python
# as the last step we create a binary image 
# to facilitate working with the image later on

ret, bin_image = cv2.threshold(filtered,
                          130, # bound below which all is equal to 0
                          255, # max value to be applied all that is above the bound
                          cv2.THRESH_BINARY
                         )

plt.imshow(bin_image, cmap='gray')
```

### Low Pass Filters and Gaussian Blur

```python
# apply gaussian blur to gray scaled image
gray_blur = cv2.GaussianBlur(image, # source image
                             (5,5), # size of the kernel
                             # bigger the kernel, bigger the blur
                             # it needs to be in odd dimensions
                             # so that we can center each of the pixels 
                             # in our image
                             0 # standard deviation
                            )
# plotting images side by side

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.set_title("original gray", color="white")
ax1.imshow(image, cmap='gray')

ax2.set_title("blur gray", color="white")
ax2.imshow(gray_blur, cmap="gray")
```

```python
# Let's see how edge detection works after all the bluring
filtered_blur = cv2.filter2D(gray_blur, # input image
                        -1, # type of output image -1 means the same type as 
                        # the input
                        conv_kernel_x # the kernel we created
                       )
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(20,10))

ax1.set_title("Edge detection no blur", color="white")
ax1.imshow(filtered, cmap='gray')

ax2.set_title("edge detection with blur", color="white")
ax2.imshow(filtered_blur, cmap="gray")
```

```python
# Here is an application of canny edge detector
hyst_lower = 250 # lower bound for hysterisis thresholding
hyst_upper = 255 # ratio should be 1/2, or 1/3.

detected_edges = cv2.Canny(image, # input image
                           hyst_lower,
                           hyst_upper
                          )
fig, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(30,10))

ax1.set_title("Edge detection no blur", color="white")
ax1.imshow(filtered, cmap='gray')

ax2.set_title("edge detection with blur", color="white")
ax2.imshow(filtered_blur, cmap="gray")


ax3.set_title("Canny detector", color="white")
ax3.imshow(detected_edges, cmap="gray")
```
